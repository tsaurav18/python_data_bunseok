````python
# 1. 캔버스 생성, 설정
# figsize=(가로 크기, 세로 크기)
plt.figure(figsize=(5, 3)) # 가로 5인치, 세로 3인치

# 2. 차트 생성
x = np.arange(0, 12, 0.01)
y = np.sin(x)
# plt.plot(x축데이터, y축데이터)
plt.plot(x, y, label='Sin') # 사인파 시각화
plt.plot(x, np.cos(x), label='Cos') # 코사인파 시각화

# 3. 옵션 설정
plt.grid() # 격자 생성
plt.title('Wave') # 캔버스 제목
plt.xlabel('time') # x축 제목 설정
plt.ylabel('amp') # y축 제목 설정
plt.xlim(0, 4) # x축 범위 설정(0~4로 설정)
plt.ylim(-0.5, 1) # y축 범위 설정(-0.5~1로 설정)
plt.legend() # label 기반으로 범례 생성

# 4. 캔버스 출력
# 저장은 출력 전에 실행
plt.savefig('wave.png', dpi= 300)
plt.show()
````

## tips 데이터셋
#### 한 레스토랑에 방문한 고객의 정보
|total_bill|tip|sex|smoker|day|time|size|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|식사비용|팁|성별|흡연여부|방문요일|방문시간대|규모|

````python
tips = sns.load_dataset('tips')
tips
````

````python
# 데이터 분포 boxplot
#요일 별로 음식값의 분포 알아보기 
#  Q1= 25%, Q2 = 50%, Q3 = 75% max Q3 + 1.5IQR   min Q1- 1.5IQR  이상치 
sns.boxplot(data=tips, x='day', y='total_bill')  
plt.show()
#sns.시각화함수(data= 데이터프레임, x='x축컬럼',y='y축컬럼', hue='법례')
````

````python
#요일별 음식값 분포, 흡연 비흡연 나눠서 출력 
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')  
plt.show()
````

````python

sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker')  
plt.show()
````

````python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')  
sns.swarmplot(data=tips, x='day', y='total_bill', hue='smoker')  

plt.show()
````

````python
#컬럼들간의 산관관계 -산점도 scatter
#food price  and tip 상관성 파악
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
````

````python
#산점도 + 상관관계 출력
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.show()
````

````python
# 여러 변수들의 상관관계를 한꺼번에 출력
# sns.pairplot(data=df, var=['컬럼1', '컬럼2', '컬럼3', ....], kind='reg')
sns.pairplot(data=tips, vars=['total_bill', 'tip', 'size'], kind='reg')
plt.show()
````

````python
# 여러 변수들의 상관관계를 한꺼번에 출력
# sns.pairplot(data=df, var=['컬럼1', '컬럼2', '컬럼3', ....], kind='reg')
sns.pairplot(data=tips, vars=['total_bill', 'tip', 'size'], kind='reg', hue='smoker')
plt.savefig('pairplot.png', dpi=300)
plt.show()
````

## flights 데이터셋
#### 한 비행기의 월간 승객수
|year|mon|passengers|
|:---:|:---:|:---:|
|탑승년도|탑승월|탑승객수|

````python
flights = sns.load_dataset('flights')
flights
````

````python
#행- 월 열-> 년도 , 데이더 승객수
flights_pivot = flights.pivot_table(index='month',
                                   columns='year',
                                   values='passengers')
flights_pivot
````

## pivot_table의 수치의 분포를 한눈에 확인시켜주는 시각화
### heatmap
- 각 타일의 데이터를 색상으로 표현하여 분포를 한눈에 확인할 수 있다.

````python
plt.figure(figsize=(15, 12))
#sns.heatmap(pivot_table, option)
#option. annot= 값 출력 여부 , fmt= 출력형태
sns.heatmap(flights_pivot, annot=True, fmt='d')
plt.show()
````
