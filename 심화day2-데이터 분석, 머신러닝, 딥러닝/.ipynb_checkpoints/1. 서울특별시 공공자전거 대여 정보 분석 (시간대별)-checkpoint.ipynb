# 서울특별시 공공자전거 대여 정보 분석

### 사용 데이터
- **서울 열린데이터 광장** - 서울특별시 공공자전거 대여이력 정보<br>https://data.seoul.go.kr/dataList/OA-15182/F/1/datasetView.do#

```python
import pandas as pd
import numpy as np
```

```python
import os
import glob
path = './data/'

# csv 파일과 xlsx 파일 리스트 분리
csv = glob.glob(path+'*.csv')

xlsx = glob.glob(path+'*.xlsx')

# 분리된 파일 확인
csv, xlsx
```

```python
# 컬럼 이름을 미리 생성
column_name = ['자전거번호', '대여일시', '대여대여소번호', '대여대여소명',
               '거치대순번', '반납일시', '반납대여소번호', '반납대여소명',
               '반납거치대순번', '이용시간', '이용거리']

# 전체 데이터를 저장할 비어있는 데이터프레임 생성
df = pd.DataFrame(columns=column_name)
df
```

```python
# CSV 파일들을 하나씩 읽어와서 df에 병합
for file in csv:
    # 읽어온 파일을 df2에 저장.
    df2 = pd.read_csv(file, encoding='cp949')
    df2.columns = column_name
    
    df = pd.concat([df, df2])
df.shape
```

```python
# XLSX 파일 병합 - 엑셀파일이라 시간이 오래걸림.
'''
for file in xlsx:
    df2 = pd.read_excel(file)
    df2.columns = column_name
    
    df = pd.concat([df, df2])
df.shape
'''
```

```python
# 병합 파일 저장
df.to_csv('./data/rent_merge.csv', index=False, encoding='euc-kr')
```

# 데이터 프레임 가공

- 대여 정보(날짜, 시간)와 반납 정보(날짜, 시간)을 분할

```python
# 데이터 프레임 컬럼 분할
df['대여날짜'] = df['대여일시'].str.split(' ', expand = True)[0]
df['대여시간'] = df['대여일시'].str.split(' ', expand = True)[1]

df['반납날짜'] = df['반납일시'].str.split(' ', expand = True)[0]
df['반납시간'] = df['반납일시'].str.split(' ', expand = True)[1]
```

```python
df.head()
```

```python
df['대여_년'] = df['대여날짜'].str.split('-', expand = True)[0]
df['대여_월'] = df['대여날짜'].str.split('-', expand = True)[1]
df['대여_일'] = df['대여날짜'].str.split('-', expand = True)[2]

df['대여_시'] = df['대여시간'].str.split(':', expand = True)[0]
df['대여_분'] = df['대여시간'].str.split(':', expand = True)[1]
```

```python
df['반납_년'] = df['반납날짜'].str.split('-', expand = True)[0]
df['반납_월'] = df['반납날짜'].str.split('-', expand = True)[1]
df['반납_일'] = df['반납날짜'].str.split('-', expand = True)[2]

df['반납_시'] = df['반납시간'].str.split(':', expand = True)[0]
df['반납_분'] = df['반납시간'].str.split(':', expand = True)[1]
```

```python
df.head()
```
