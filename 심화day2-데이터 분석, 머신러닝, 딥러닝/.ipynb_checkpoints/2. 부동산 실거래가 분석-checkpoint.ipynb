```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
    
df_apt = pd.read_csv('./서울시 아파트 실거래가.csv')
df_loc = pd.read_csv('./법정동코드 전체자료.txt', sep='\t')
display(df_apt, df_loc)
```

```python
df_loc.columns = ['ji_code', '시도명', '시군구명']
df_loc
```

```python
df_apt = pd.merge(df_apt, df_loc, on='ji_code')
df_apt
```

```python
df_apt['시군구명']
```

```python
df_apt['price'] = df_apt['price'].str.replace(',','').astype('int')
```

```python
df_apt
```

```python
df1 = df_apt.pivot_table('price',
                     index='시군구명',
                     columns='month', aggfunc=len)
df1
```

### 한글폰트설정

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df1, cmap='RdPu')
```

```python
df2 = df_apt.pivot_table(values='price', index='시군구명',aggfunc='count')
df2
```

```python
df2.plot(kind="bar")
plt.show()
```

```python
df3 = df_apt.pivot_table(values='price', index='month', aggfunc='count')
df3
```

```python
df3.plot(kind="bar")
plt.show()
```

```python
df4 = df_apt.pivot_table(values='price',
                     index='시군구명',
                     columns='month', aggfunc=sum)
df4
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df4, cmap='RdPu')
plt.show()
```

```python
df5 = df_apt.groupby(by=['시군구명', 'dong']).agg('sum')
df5
```

```python
df6 = df_apt[df_apt.시군구명.isin(['강남구', '노원구'])]
df6
```

```python
df6_pivot = df6.pivot_table(values='price', index='시군구명', 
                            columns='month', aggfunc='sum')
df6_pivot
```

```python
plt.figure(figsize=(15,4))
sns.heatmap(df6_pivot, cmap='RdPu')
plt.show()
```
